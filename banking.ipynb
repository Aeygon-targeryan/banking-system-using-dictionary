```python
cust_data = {}
new_user_attributes = []
import random

def new_user():
    global account ,dict1 , new_user_attributes,cust_data
    account =random.randint(10000,99999)
    user ={}
    dict1 = {account : user}
    
    name = input("Enter customer name: ")
    address = input("Enter addess : ")
    govt_id = input(" govt id : ")
    amount = int(input("Enter amount : "))

    user["name"] = name
    user["address"] = address
    user["govt id"] = govt_id
    user["amount"] = amount
    dict1[account].update(user)
    new_user_attributes.append(dict1)
    
    for x in new_user_attributes:                   #extract dictionary from list 
        for key,value in x.items():
            cust_data.update({key : value})
            
    print("Account Successfully created !!")
    print("Your account number is : ",account)
    print("Please don't loose it !!")

def existing_user():
    n = int(input("Enter existing account number : "))
    if (n in cust_data):
        print(cust_data[n])
    else:
        print("Not found")
        print()
        
    print("WELCOME HORIZON BANK !!")
        
    print("choice 1 : display amount \nchoice 2 :  withdrawal amount from the available balance \nchoice 3 :  deposit amount to the available balance ")
    user_choice = input("user input in 1,2,3 :  ")
    
    while user_choice  in ['1','2','3']:
        if (user_choice == "1"):
            n = int(input("Enter account number : "))
            return cust_data[n]["amount"]
        
        
        elif user_choice == "2":
            amount =int(input("Enter amount to withdrawl in bank : "))
            available_balance = cust_data[n]["amount"]    
            if (amount > available_balance):
                
                return "you have insufficient balance !!"
            
            else:
                amt = available_balance - amount
                cust_data[n]["amount"] = amt
                return available_balance - amount
            
            
        elif user_choice == "3":
            amount = int(input("Enter amount to deposit in bank : "))
            available_balance = cust_data[n]["amount"]
            amt = available_balance + amount
            cust_data[n]["amount"] = amt
            return  available_balance + amount
        
        
        else : 
            return "Dang se Daal"

while True:
    print("only enter 1 and 2 number only ")
    user_choice = input("Enter number : ")
    while user_choice in ['1','2','3']:
        print()
        print("Enter 1 if you are a new customer.\nEnter 2 if you are an existing customer.\nEnter 3 to terminate the application.\n")
        user_choice = input()
        
        if (user_choice == "1"):
            print(new_user())
            
            
        elif (user_choice=="2"):
            print(existing_user())
        
        
        else:
            print('Thank you, for banking with us!')
            break
    if user_choice not in  [1,2,3]:
        break
```
